# Выявление закономерностей, определяющих успешность игры.

# Описание проекта

Вы работаете в интернет-магазине «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Вам нужно выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.
В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

# Описание данных

Name — название игры,
Platform — платформа,
Year_of_Release — год выпуска,
Genre — жанр игры,
NA_sales — продажи в Северной Америке (миллионы проданных копий),
EU_sales — продажи в Европе (миллионы проданных копий),
JP_sales — продажи в Японии (миллионы проданных копий),
Other_sales — продажи в других странах (миллионы проданных копий),
Critic_Score — оценка критиков (максимум 100),
User_Score — оценка пользователей (максимум 10),
Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board).

## Подготовка данных

### Замена названий столбцов (приведение к нижнему регистру)

```python
df.columns = df.columns.str.lower()
df.info()
```

### Аббревиатура 'tbd' в столбцах с рейтингом.

```python
# заменяем tbd на NaN
df['user_score'] = df['user_score'].replace('tbd', np.nan)
```

Если загуглить аббревиатуру "tbd", то мы узнаем, что это соответствует значению "Подлежит уточнению". То есть тот же NaN. Заменим на него.

### Обработка пропусков.

```python
# удаляем строки с пропусками в колонках, где количество пропущенных значений мало
for column in ['year_of_release', 'name', 'genre']:
    df.dropna(subset = [column], inplace = True)
```

```python
# заменяем для столба с рейтингом пропущенные значения на rating undefined
df['rating'] = df['rating'].replace(np.nan, 'rating undefined')
```

### Преобразование данных в нужные типы.

```python
# заменяем тип данных в колонке year_of_release на int
df['year_of_release'] = df['year_of_release'].astype(int)
```

```python
# тип данных в колонке user_score изменяем с object на float
df['user_score'] = df['user_score'].astype('float')
```

```python
# проверяем, целесообразно ли названия платформ загонять в category
df['platform'].unique()
```

```python
# для значений в колонках 'platform', 'genre', 'rating' меняем тип на category
for column in ['platform', 'genre', 'rating']:
    df[column] = df[column].astype('category')
```

### Рассчет суммарных продаж во всех регионах.

```python
# добавляем сумму продаж всех регионов, как отдельный столбец.
df['sales_total'] = df['na_sales'] + df['eu_sales'] + df['jp_sales'] + df['other_sales']
```

```python
#df['score_total'] = df['user_score'] + df['critic_score']
```

Были преобразованы данные в нужные типы для удобства проведения дальнейших расчетов и экономии используемой памяти, обработны пропуски, посчитаны суммарные продажи во всех регионах. Найдено объяснение значению 'tbd' в столбцах с рейтингом и заменено на Nan. ## Проверка гипотез

### Средние пользовательские рейтинги платформ Xbox One и PC одинаковые

```python
# избавимся, наконец, от значений NaN в нашей таблице.
df = df.dropna(subset=['user_score'])
```

Нулевая гипотеза: средние пользовательские рейтинги платформ Xbox One и PC одинаковые.
Альтернативная гипотеза: средние пользовательские рейтинги платформы Xbox One отличаются от рейтингов PC.
Пороговое значение альфа устанавливаем равным 5%.
Критерий для проверки гипотез: t-Критерий Стьюдента.
Устанавливаем equal_var = False, так как дисперсии выборок нельзя считать равными.

```python
# проверяем нулевую гипотезу о равенстве пользовательских рейтингов платформ Xbox One и PC
xone = df.query('platform == "XOne"')['user_score']
pc = df.query('platform == "PC"')['user_score']
alpha = .05
result = st.ttest_ind(xone, pc, equal_var = False)
print('p-value равен', result.pvalue)
if  result.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не удалось отвергнуть нулевую гипотезу')
```

```python
# смотрим среднее для рейтингов Xbox
df.query('platform == "XOne"')['user_score'].mean()
```

```python
# смотрим среднее для рейтингов PC
df.query('platform == "PC"')['user_score'].mean()
```

### Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

Нулевая гипотеза: средние пользовательские рейтинги жанров Action и Sports одинаковые.
Альтернативная гипотеза: средние пользовательские рейтинги жанров Action и Sports разные.
Пороговое значение альфа устанавливаем равным 5%.
Критерий для проверки гипотез: t-Критерий Стьюдента.
Устанавливаем equal_var = False, так как дисперсии выборок нельзя считать равными.

```python
# проверяем нулевую гипотезу о равенстве средних пользовательских рейтингов жанров Action и Sports
action_genre = df.query('genre == "Action"')['user_score']
sports_genre = df.query('genre == "Sports"')['user_score']
alpha = .05
result = st.ttest_ind(action_genre, sports_genre, equal_var = False)
print('p-value равен', result.pvalue)
if  result.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не удалось отвергнуть нулевую гипотезу')
```

```python
# смотрим среднее для рейтингов Action
df.query('genre == "Action"')['user_score'].mean()
```

```python
# смотрим среднее для рейтингов Sports
df.query('genre == "Sports"')['user_score'].mean()
```

### Вывод
Мы провели два теста для подтверждения гипотез о равенстве пользовательских рейтингов платформ Xbox One и PC и о различиях у средних пользовательских рейтингов жанров Action и Sports. При этом первая нулевая гипотеза была отвергнута с малым шансом на совпадение средних выборок, а вторая альтернативная гипотеза подтверждается. То есть средние пользовательские рейтинги платформ Xbox One и PC схожи, а средние пользовательские рейтинги жанров Action и Sports различаются (у Action рейтинги выше).
Для проверки был использован t-Критерий Стьюдента, так как он подходит для проверки равенства средних значений в двух выборках.
